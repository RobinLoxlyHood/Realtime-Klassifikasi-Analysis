```python
import pymysql
import pandas as pd

def read_mysql_table(table, host='localhost', user='root', password='', database='klasifikasi_sentimen'):
    # Establish a connection to the MySQL database
    connection = pymysql.connect(
        host=host,
        user=user,
        password=password,
        database=database
    )
    
    # Create a cursor object to execute SQL queries
    cursor = connection.cursor()
    
    query = f"SELECT * FROM {table}"
    cursor.execute(query)
    result = cursor.fetchall()
    
    # Convert the result to a Pandas DataFrame
    df = pd.DataFrame(result)
    
    # Assign column names based on the cursor description
    df.columns = [column[0] for column in cursor.description]
    
    # Close the cursor and the database connection
    cursor.close()
    connection.close()
    
    return df
```

```python
table_name = 'hasil_scraping'
df = read_mysql_table(table_name)

# Lakukan manipulasi data atau analisis dengan DataFrame yang dihasilkan
print(df.head())
```
